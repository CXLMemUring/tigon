```python
import altair as alt
import plotly.express as px
import polars as pl

from parse import parse

ACCESSES = [
    [
        "local_access",
        "local_cxl_access",
    ],
    [
        "remote_access",
        "remote_access_with_req",
    ]
]

COLORS = ["blue", "orange"]

data_movement = parse("../../results/pasha/202412060808/data-movement")
data_movement_tpcc = (
    data_movement.lazy()
        .filter(pl.col("benchmark") == "tpcc")
        .sort(pl.col("max_migrated_rows_size", "neworder_dist"))
        .collect()
)

def plot_stack(groups):
    return alt.vconcat(*[
        alt.Chart(data_movement_tpcc)
        .mark_area()
        .transform_fold(fold=group)
        .encode(x="neworder_dist", y="value:Q", color="key:N")
        .facet(
            "max_migrated_rows_size",
        )
        for group in groups
    ])

plot_stack(ACCESSES)
```

```python
MOVES = [["data_move_in", "data_move_out"]]

plot_stack(MOVES)
```

```python
LATENCIES = [
    [
        "wal_group_commit_lat_p50",
        "wal_group_commit_lat_p95",
        "wal_group_commit_lat_p99",
    ]
]

plot_stack(LATENCIES)
```

```python
SWCC = [
    [
        "num_clflush",
        "num_clwb",
    ],
    [
        "num_cache_hit",
        "num_cache_miss",
    ],
]

plot_stack(SWCC)
```
